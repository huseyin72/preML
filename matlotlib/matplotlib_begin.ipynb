```python
ageList = np.random.randint(10,70,15)
weigthList = np.random.randint(40,130,15)
```

```python
numpyArray1 = np.linspace(0,10,20)
numpyArray2 = numpyArray1**2
```

```python
plt.plot(numpyArray1,numpyArray2,"b--") #b*-
```

```python
plt.subplot(1,2,1) #One line two column. this represesnts column1
plt.plot(numpyArray1,numpyArray2,"r*-")
plt.subplot(1,2,2)#This represents column 2
plt.plot(numpyArray2,numpyArray1,"g--")
```

## Create own figure

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.1 ,0.5,0.2,0.3]) #0.2 width 0.3 height
figureAxes.plot(numpyArray1,numpyArray2,"b")
figureAxes.set_xlabel("X axes")
figureAxes.set_ylabel("Y axes")
figureAxes.set_title("its title")
```

```python
myFigure2 = plt.figure()
figureAxes1 = myFigure2.add_axes([0.1 ,0.1,0.9,0.9]) #0.2 width 0.3 height
figureAxes1.plot(numpyArray1,numpyArray2,"b")
figureAxes1.set_xlabel("X axes")
figureAxes1.set_ylabel("Y axes")
figureAxes1.set_title("its title")

figureAxes2 = myFigure2.add_axes([0.3 ,0.6,0.3,0.3]) #0.2 width 0.3 height
figureAxes2.plot(numpyArray1,numpyArray2,"b")
figureAxes2.set_xlabel("X axes")
figureAxes2.set_ylabel("Y axes")
figureAxes2.set_title("its title")
```

```python
newFigure = plt.figure()
newAxes = newFigure.add_axes([0.1,0.1,0.9,0.9])
newAxes.plot(numpyArray1,numpyArray2,label="NumpyArray**2")

newAxes.legend()
```
